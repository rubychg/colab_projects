```python
import nltk
from nltk.corpus import PlaintextCorpusReader

corpus_root = "/content/drive/MyDrive/Colab Notebooks/TXT"
A = PlaintextCorpusReader(corpus_root, ".*")

books = A.fileids()
print(books)
```

```python
import nltk
testText = A.words('Definition Final Draft.txt')
single = nltk.text.Text(testText)

findList = ["first", "firstly", "secondly", "also", "last", "finally"]
single.dispersion_plot(findList)
```

```python
books = A.fileids()

cfd = nltk.ConditionalFreqDist((fileid, word)
    for fileid in books
    for word in A.words(fileid)
    )

findList = ["first", "firstly", "secondly", "also", "last", "finally"]
cfd.tabulate(conditions=books, samples=findList)
```

```python
cfd = nltk.ConditionalFreqDist(
      (target, fileid[:5])
      for fileid in A.fileids()
      for w in A.words(fileid)
      for target in ['should', 'it', 'i']
      if w.lower() == target)
cfd.plot()
```

```python
import nltk
from wordcloud import STOPWORDS

dataset = A.raw('Definition Final Draft.txt')

tokens = dataset.split(' ')
clean_tokens = tokens[:]

for token in tokens:
  if token.lower() in STOPWORDS:
    clean_tokens.remove(token)

freq = nltk.FreqDist(clean_tokens)

freq.plot(20,cumulative=False)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

file_text = open ('/content/drive/MyDrive/Colab Notebooks/TXT/Definition Final Draft.txt', 'r').read()

wordcloud = WordCloud(
    background_color="white",
    width=2000,
    height=1000,
    margin=10,
).generate(file_text)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

f = open ('/content/drive/MyDrive/Colab Notebooks/TXT/Definition Final Draft.txt', 'r').read()

stop = list(STOPWORDS)

wordcloud = WordCloud(
    background_color="#fff",
    width=1500,
    height=1000,
    margin=2,
    stopwords=stop
).generate(f)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
